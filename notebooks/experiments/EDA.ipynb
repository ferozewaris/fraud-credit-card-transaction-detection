```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
# File location and type
file_location = "/FileStore/tables/fraud_detection/transaction_data_updated.csv"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)
```

```
cols_plt_dst = pdf.select_dtypes(include=['object', 'bool']).drop(['currentExpDate', 'merchantName', 'isFraud'], axis=1)
for col in cols_plt_dst:
      sns.countplot(y=col, data=pdf)
      plt.show()
```

```
plt.style.use("ggplot")
cols_plt_dst = pdf.select_dtypes(include=['object', 'bool']).drop(['currentExpDate', 'merchantName'], axis=1)
for column in cols_plt_dst:
    plt.figure(figsize=(20,4))
    plt.subplot(121)
    pdf[column].value_counts().plot(kind="bar")
    plt.xlabel(column)
    plt.ylabel("count")
    plt.title(column)
```

```
plt.style.use("ggplot")
for column in cols_plt_dst:
    plt.figure(figsize=(20,4))
    plt.subplot(121)
    sns.countplot(pdf[column], hue=pdf["isFraud"])
    plt.title(column)    
    plt.xticks(rotation=90)
```

```
# plotting histogram for each numerical variable
plt.style.use("ggplot")
for column in ['creditLimit', 'availableMoney', 'transactionAmount', 'currentBalance']:
    plt.figure(figsize=(20,4))
    plt.subplot(121)
    sns.distplot(pdf[column], kde=True)
    plt.title(column)
```
